3. **Create Time based features**

```python
data = Data_manager.change_to_date('Order Date')

data = Data_manager.create_Time_Based_features()

# --- Plotting ---
# Combine "Month" and "Year" into a datetime column
data['Date'] = pd.to_datetime(data['Year'].astype(str) + '-' + data['Month'].astype(str), format='%Y-%m')

# Group by the combined date and calculate the sum of sales
df_summary = data.groupby('Date')['Sales'].sum().reset_index()

# Create a line plot
plt.figure(figsize=(10, 5))
plt.plot(df_summary['Date'], df_summary['Sales'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Over Time-Monthly View')
plt.grid(True)
plt.show()
```

```python
#We need to check if there are any missing dates on the daily trends and populate those dates with 0 sales to ensure proper decomposition
# on the daily view. Also, some models will not work well if the date trend is not consistent.


# Define the date range
date_range = pd.date_range(start=data['Order Date'].min(), end=data['Order Date'].max())

# Find missing dates
missing_dates = date_range.difference(data['Order Date'])
print(f"Missing dates: {len(missing_dates)}")

# Add missing dates to the DataFrame with Sales number of 0
missing_data = {
    "Order Date": missing_dates,
    "Sales": [0] * len(missing_dates)
}
missing_df = pd.DataFrame(missing_data)

# Concatenate the missing data with the original DataFrame
data = pd.concat([data, missing_df], ignore_index=True)

# Extract the "Day," "Month," and "Year" from the "Order Date" column
data['Day'] = data['Order Date'].dt.day
data['Month'] = data['Order Date'].dt.month
data['Year'] = data['Order Date'].dt.year

# Sort the DataFrame by "Order Date"
data = data.sort_values(by='Order Date')


data.head()
```
